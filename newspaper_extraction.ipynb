The first part of the project involves scanning through screenshots of Straits Times articles on the Racial Riots to identify locations where violent incidents occurred.

```python
f = open("newspaper.txt","x") #Creates a .txt file where the OCR text will be stored
```

The first part consists of using Pytesseract to perform optical character recognition on the given screenshots of the newspapers and digitise their contents for later analysis.

```python
#Performs OCR on the given screenshots of newspapers and writes them to the previously created file
!pip install pytesseract

import pytesseract
import os

directory_path = "/content/newspapers"
directory_files = os.listdir(directory_path)

newspaper_text = []
for img_file in directory_files:
  img = os.path.join(directory_path, img_file)
  text = pytesseract.image_to_string(img)
  newspaper_text.append(text)

with open("newspaper.txt", "a") as f:
  f.writelines(newspaper_text)
```

After cleaning up the scanned texts, the next step involves performing natural language processing, namely named entity recognition to identify any named places in the texts.

```python
import spacy
nlp = spacy.load("en_core_web_sm")

with open("compiled.txt", "r") as f:
  reports = f.read().splitlines()

#Performs nanmed entity recognition on the text
entities = []
for report in reports:
  doc = nlp(report)
  report_ner = [{"Text": ent.text, "Label": ent.label_} for ent in doc.ents]
  entities.append(report_ner)
```

After looking at the results, I noticed that some locations (e.g. North Bridge Road) were identified as facilities or locations even though they should have been identified as geopolitical entities.

```python
#Removes everything that is not a place (An entity that is labelled "LOC", "FAC" or "GPE")
places = []
spacy_ner_except_wanted = ["CARDINAL", "DATE", "EVENT", "LANGUAGE", "LAW", "MONEY", "NORP", "ORDINAL", "ORG", "PERCENT", "PERSON", "PRODUCT", "QUANTITY", "TIME", "WORK_OF_ART"]
for entity in entities:
  for ner in entity:
    if ner["Label"] not in spacy_ner_except_wanted: #In retrospective, if ner["Label"] == "LOC" or ner["Label"] == "FAC" or ner["Label"] == "GPE" would be easier than what I did
      places.append(ner["Text"])

with open("locations.txt","w") as f:
  f.writelines("\n".join(places))
```

After cleaning up the results, the next step is to count the number of time each location appears.

```python
#Counts the number of times each location appears
from collections import Counter

with open("locations_cleaned.txt","r") as f:
  all_locations = f.read().splitlines()
loc_count = Counter(all_locations)
print(loc_count)
```
